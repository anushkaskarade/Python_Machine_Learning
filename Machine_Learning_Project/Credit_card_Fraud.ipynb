```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
#load the data in pandas file
Credit_card_data=pd.read_csv('Credit_card_fraud_data.csv')
Credit_card_data.head()
```

```python
#Removing unnecessary columns from dataset
Credit_card_data=Credit_card_data.drop(columns=['GENDER'],axis=1)
Credit_card_data=Credit_card_data.drop(columns=['CAR'],axis=1)
Credit_card_data=Credit_card_data.drop(columns=['INCOME_TYPE'],axis=1)
Credit_card_data=Credit_card_data.drop(columns=['EDUCATION_TYPE'],axis=1)
Credit_card_data=Credit_card_data.drop(columns=['FAMILY_TYPE'],axis=1)
Credit_card_data=Credit_card_data.drop(columns=['HOUSE_TYPE'],axis=1)
Credit_card_data
```

```python

Credit_card_data.info()
```

```python
Credit_card_data.isnull().sum()
```

```python
# Adding a new column using if-else
Credit_card_data['Class'] = [0 if x == 'Y' else 1 for x in Credit_card_data['REALITY']]
print(Credit_card_data)
```

```python
Credit_card_data=Credit_card_data.drop(columns=['REALITY'],axis=1)
Credit_card_data
```

```python
#Printing the count of normal transaction and fraudulent transaction
Credit_card_data['Class'].value_counts()
```

```python
#Data is imbalanced.
#Seperating the data for analysis
legit=Credit_card_data[Credit_card_data.Class==0]
Fraud=Credit_card_data[Credit_card_data.Class==1]
print(legit.shape)
print(Fraud.shape)
```

```python
#The data is unbalanced
#Statistical measures of data
legit.INCOME.describe()
```

```python
Fraud.INCOME.describe()
```

```python
#Number of fraudulent transaction=6492
legit_sample=legit.sample(n=6452)
```

```python
new_dataset=pd.concat([legit_sample,Fraud],axis=0)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
#Splitting the data into targets and features
#X=features
#Y=labels
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
print(X) 
```

```python
print(Y)
```

```python
#Split the data into Training and Testing data
#20% for testing and 80% data for testing 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
#training the logistic Regression Model with Training data
#model=LogisticRegression(max_iter=1000)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, max_depth=10)





```

```python
model.fit(X_train,Y_train)
```

```python
#Model Evaluation
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data: ',training_data_accuracy)
```

```python
#Accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on test data:',test_data_accuracy);
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
Y_pred=model.predict(Y_test)
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```
